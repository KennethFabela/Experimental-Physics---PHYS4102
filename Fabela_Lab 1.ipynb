## Gausian Fit - HD1 - Peak 1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def GausFit(x, xbar, sigma, A):
    return A * np.exp(-(x - xbar)**2 / (2 * sigma**2))

def plot_graph_with_gaussian_fit(filename, fit_range):
    # Read the numbers from the file
    with open(filename, 'r') as file:
        numbers = [float(line.strip()) for line in file.readlines()]

    # Create x-axis values (assuming consecutive integers starting from 1)
    x_values = list(range(1, len(numbers) + 1))

    # Decrease the scale of the x-axis by a factor of 10
    x_values_scaled = [x / 10 for x in x_values]

    # Set figure size to elongate the graph
    plt.figure(figsize=(12, 7))  # Adjust width and height as needed

    # Plot the graph with black color for the data line
    plt.plot(x_values_scaled, numbers, color='black', label='Data')
    plt.title('Hydrogen Emission Spectrum 1 Voltage vs. Time With Gaussian Fit - Peak 1')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Voltage (V)')
    plt.grid(True)

    # Set x-axis range from 0 to 1100 with steps of 100
    plt.xticks(range(0, 1155, 50))

    # Restrict x-values for fitting
    fit_indices = [i for i, val in enumerate(x_values_scaled) if fit_range[0] <= val <= fit_range[1]]
    fit_x_values = [x_values_scaled[i] for i in fit_indices]
    fit_numbers = [numbers[i] for i in fit_indices]

    # Perform Gaussian fit
    p0 = (np.mean(fit_x_values), np.std(fit_x_values), max(fit_numbers))
    params, cov = curve_fit(GausFit, fit_x_values, fit_numbers, p0=p0)
    x_bar = params[0]
    fit_values = GausFit(fit_x_values, *params)
    plt.plot(fit_x_values, fit_values, 'r-', label='Gaussian Fit')

    plt.xlim(fit_range)  # Set x-axis range
    plt.legend()
    plt.show()

    print("x_bar:", x_bar,"s")

filename = 'Lab 1/processed_hydrogen_data_1.txt'  # Path of text file
fit_range = (281.095, 283.4)  # Range of x-values for Gaussian fit
plot_graph_with_gaussian_fit(filename, fit_range)
```

## New calculation for m and b using Gaussian fit:

Peak 1: 486.1nm at 282.1962s

Peak 2: 656.2nm at 824.2112s

#### Calculating m:

Change in wavelength: 

$656.2nm - 486.1nm = 170.1nm$

Time difference between peaks:

$824.2112s - 282.1962s = 542.015s$

$m = \frac{wavelength}{time} = nm/s$

$m = \frac{170.1nm}{542.015s} = 0.3138nm/s$

### m = 0.3138nm/s

### Calculating b:

$486.1nm =  (0.3138nm/s)(282.1962s) + b$

$b = 486.1nm - (0.3138nm/s)(282.1962s)$

$b = 486.1nm - 88.5532nm$

## Gaussian Fit - HE1 - Peak 1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def GausFit(x, xbar, sigma, A):
    return A * np.exp(-(x - xbar)**2 / (2 * sigma**2))

def plot_graph_with_gaussian_fit(filename, fit_range):
    # Read the numbers from the file
    with open(filename, 'r') as file:
        numbers = [float(line.strip()) for line in file.readlines()]

    # Create x-axis values (assuming consecutive integers starting from 1)
    x_values = list(range(1, len(numbers) + 1))

    # Decrease the scale of the x-axis by a factor of 10
    x_values_scaled = [x / 10 for x in x_values]

    # Set figure size to elongate the graph
    plt.figure(figsize=(12, 7))  # Adjust width and height as needed

    # Plot the graph with black color for the data line
    plt.plot(x_values_scaled, numbers, color='black', label='Data')
    plt.title('Helium Emission Spectrum 1 With Gaussian Fit - Peak 1')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Voltage (V)')
    plt.grid(True)

    # Set x-axis range from 0 to 1100 with steps of 100
    plt.xticks(range(0, 1155, 50))

    # Restrict x-values for fitting
    fit_indices = [i for i, val in enumerate(x_values_scaled) if fit_range[0] <= val <= fit_range[1]]
    fit_x_values = [x_values_scaled[i] for i in fit_indices]
    fit_numbers = [numbers[i] for i in fit_indices]

    # Perform Gaussian fit
    p0 = (np.mean(fit_x_values), np.std(fit_x_values), max(fit_numbers))
    params, cov = curve_fit(GausFit, fit_x_values, fit_numbers, p0=p0)
    x_bar = params[0]
    fit_values = GausFit(fit_x_values, *params)
    plt.plot(fit_x_values, fit_values, 'r-', label='Gaussian Fit')

    plt.xlim(fit_range)  # Set x-axis range
    plt.legend()
    plt.show()

    print("x_bar:", x_bar,"s")

filename = 'Lab 1/processed_helium_data_1.txt'  # Path of text file
fit_range = (153, 156.5)  # Range of x-values for Gaussian fit
plot_graph_with_gaussian_fit(filename, fit_range)
```

#### Calculating m:

Change in wavelength: 

$587.56nm - 447.1nm = 140.5nm$

Time difference between peaks:

$606.6498s - 154.836s = 451.8138s$

$m = \frac{wavelength}{time} = nm/s$

$m = \frac{140.5nm}{451.8138s} = 0.31097nm/s$

### m = 0.31097nm/s

### Calculating b:

$587.56nm =  (0.31097nm/s)(606.6498s) + b$

$b = 587.56nm - (0.31097nm/s)(606.6498s)$

$b = 587.56nm - 188.6499nm$

$b = 398.91nm$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def GausFit(x, xbar, sigma, A):
    return A * np.exp(-(x - xbar)**2 / (2 * sigma**2))

def plot_graph_with_gaussian_fit(filename, fit_range):
    # Read the numbers from the file
    with open(filename, 'r') as file:
        numbers = [float(line.strip()) for line in file.readlines()]

    # Create x-axis values (assuming consecutive integers starting from 1)
    x_values = list(range(1, len(numbers) + 1))

    # Decrease the scale of the x-axis by a factor of 10
    x_values_scaled = [x / 10 for x in x_values]

    # Set figure size to elongate the graph
    plt.figure(figsize=(12, 7))  # Adjust width and height as needed

    # Plot the graph with black color for the data line
    plt.plot(x_values_scaled, numbers, color='black', label='Data')
    plt.title('Argon Emission Spectrum 2 With Gaussian Fit - Peak 1')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Voltage (V)')
    plt.grid(True)

    # Set x-axis range from 0 to 1100 with steps of 100
    plt.xticks(range(0, 1155, 50))

    # Restrict x-values for fitting
    fit_indices = [i for i, val in enumerate(x_values_scaled) if fit_range[0] <= val <= fit_range[1]]
    fit_x_values = [x_values_scaled[i] for i in fit_indices]
    fit_numbers = [numbers[i] for i in fit_indices]

    # Perform Gaussian fit
    p0 = (np.mean(fit_x_values), np.std(fit_x_values), max(fit_numbers))
    params, cov = curve_fit(GausFit, fit_x_values, fit_numbers, p0=p0)
    x_bar = params[0]
    fit_values = GausFit(fit_x_values, *params)
    plt.plot(fit_x_values, fit_values, 'r-', label='Gaussian Fit')

    plt.xlim(fit_range)  # Set x-axis range
    plt.legend()
    plt.show()

    print("x_bar:", x_bar,"s")

filename = 'Lab 1/processed_argon_data_2.txt'  # Path of text file
fit_range = (61.25, 72.25)  # Range of time x-values for Gaussian fit
plot_graph_with_gaussian_fit(filename, fit_range)
```
