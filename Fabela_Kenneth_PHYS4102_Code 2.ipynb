### c) Creating a running average of the data and plotting both the raw data and running average on the same plot:

```python
from pylab import scatter, xlabel, ylabel, xlim, show, title, linspace, sin
import matplotlib.pyplot as plt

#data = loadtxt("sunspots.txt", float)
x_raw = data[:, 0]
y_raw = data[:, 1]

# Raw data scatter plot
plt.scatter(x_raw, y_raw, color='blue', marker='o', label='Raw Data')

# Average calculation
x_values = []  # List to store x values
y_values = []  # List to store y values

for index in range(len(y_raw)):
    s = 0
    count = 0
    for ar in range(-2, 2 + 1):
        if 0 <= index + ar < len(y_raw):
            s += y_raw[index + ar]
            count += 1
    avg = s / count
    x = index  # Assigning the x value as the current index
    x_values.append(x)  # Append x value to the list
    y_values.append(avg)  # Append y value to the list

# Moving average scatter plot
plt.scatter(x_values, y_values, color='red', marker='o', label='Running Average')

# Adds labels and title
plt.xlabel('Time (months)')
plt.ylabel('Number of Sunspots')
plt.title('Sunspots Over 50 Years Since 1749')

# This is the range of the first five peaks
plt.xlim(0, 600)
plt.ylim(0, 250)

# Add legend
plt.legend()

# Add grid
plt.grid(True)

# Show plot
plt.show()
```

# 2) Creating 1D interference between two sine waves. Then follow the 2D interference example.
### a) 1D interference between two sine waves:

```python
import numpy as np
import matplotlib.pyplot as plt

# Defines the time axis
t = np.linspace(0, 1, 1000)  # 1 second, 1000 points

# Defines frequencies
f1 = 100  # Hz
f2 = 105  # Hz

# Generates the waves
wave1 = np.sin(2 * np.pi * f1 * t)
wave2 = np.sin(2 * np.pi * f2 * t)

# Plots the waves on separate axes
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, wave1, label='Wave 1 (100 Hz)')
plt.plot(t, wave2, label='Wave 2 (105 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Two Sine Waves')
plt.legend()

# Adds the two waves together
combined_wave = wave1 + wave2

plt.subplot(2, 1, 2)
plt.plot(t, combined_wave, color='red', label='Combined Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

### b) Following Example 3.1 in the textbook creating a 2D interference plot:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the interference pattern
wavelength, xi0, separation, side, points = 5.0, 1.0, 20.0, 100.0, 500

# Calculates the wave number
k = 2 * np.pi / wavelength

# Calculates spacing between grid points
spacing = side / points

# Coordinates of first source
x1, y1 = side / 2 + separation / 2, side / 2

# Coordinates of second source
x2, y2 = side / 2 - separation / 2, side / 2

# Creates array and generates interference pattern
xi = np.fromfunction(lambda i, j: xi0 * (np.sin(k * np.sqrt((j * spacing - x1)**2 + (i * spacing - y1)**2)) + np.sin(k * np.sqrt((j * spacing - x2)**2 + (i * spacing - y2)**2))), (points, points))

# Displays interference pattern
plt.imshow(xi, origin="lower", extent=[0, side, 0, side], cmap='gray')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

# Reads the CSV file
df = pd.read_csv("CovidData.csv")

# Converts the date column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Filters data from March 4th, 2020 onwards
start_date = datetime(2020, 3, 4)
df = df[df['Date'] >= start_date]

# Calculates the number of days since March 4th, 2020
df['Days'] = (df['Date'] - start_date).dt.days

# Filters data for number of deaths between 10 to 50 days
filtered_df = df[(df['Days'] >= 10) & (df['Days'] <= 50)]

# Plots data
plt.figure(figsize=(10, 6))

# Plots the number of deaths
plt.plot(filtered_df['Days'], filtered_df['NDeaths'], label='Number of Deaths (COVID-19)', color='red')

# Defines the range for x values for the exponential curve
x_values = np.linspace(10, 50, 100)

# Defines the exponential function parameters
a = 47  # Adjust the parameters as needed
b = 0.07
c = -90

# Calculates the corresponding y values using the exponential function
y_values = a * np.exp(b * x_values) + c

# Plots the exponential curve
plt.plot(x_values, y_values, label='Exponential Fit')

# Adds labels and title
plt.xlabel('Days since March 4th, 2020')
plt.ylabel('Number of Deaths')
plt.title('Number of Deaths and Exponential Fit')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

# Reads the CSV file
df = pd.read_csv("CovidData.csv")

# Converts the date column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Filters data from March 4th, 2020 onwards
start_date = datetime(2020, 3, 4)
df = df[df['Date'] >= start_date]

# Calculates the number of days since March 4th, 2020
df['Days'] = (df['Date'] - start_date).dt.days

# Filters data for number of deaths between day 250 to day 300
filtered_df = df[(df['Days'] >= 250) & (df['Days'] <= 300)]

# Plots data
plt.figure(figsize=(10, 6))

# Plots the number of deaths
plt.plot(filtered_df['Days'], filtered_df['NDeaths'], label='Number of Deaths (COVID-19)', color='red')

# Defines the range for x values for the exponential curve
x_values = np.linspace(250, 300, 100)

# Defines the exponential function parameters to fit the desired range
a = 650  # Adjust the parameters as needed
b = 0.048
c = 17350

# Calculates the corresponding y values using the exponential function
y_values = a * np.exp(b * (x_values - 250)) + c

# Plots the exponential curve
plt.plot(x_values, y_values, label='Exponential Fit')

# Adds labels and title
plt.xlabel('Days since March 4th, 2020')
plt.ylabel('Number of Deaths')
plt.title('Number of Deaths and Exponential Fit')
plt.legend()
plt.grid(True)

plt.show()
```
